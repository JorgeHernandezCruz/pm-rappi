```python
#se caraga el documento de subcategorias RT (Se baja la tabla con porcentajes) como csv y se crea el dataframe para operar con el
#Leemos el csv que extraemos de Control Panel solo 16 Semanas LW y This Week
dfrtp = pd.read_csv("SUBRT.csv")
dfrtp
```

```python
#Agrupamos el Dataframe por semana y creamos el diccionario con Key definida por numero de dia 
#Para RT
dfs_by_day_rtp = {day: subset for day, subset in dfrtp.groupby('Day')}
print(type(dfs_by_day_rtp))
print(dfs_by_day_rtp)
```

```python
#Imprimimos el Dataframe para corroborar que se secciono por dia
print(dfs_by_day_rtp[16]) #Key de numero de Dia de semana  Cambiar
print(dfs_by_day_rtp[23])
```

```python
#Colocar for que llene las categorias faltantes para que el siguiente bloque de codigo complete la fecha faltante 
#Generar las categorias faltantes y asignarles 0 en el nuevo DF
# Reemplazar coma por punto y eliminar el porcentaje
dfrtp["Cancel_rate"] = dfrtp["Cancel_rate"].str.replace(",", ".").str.replace(" %", "").astype(float)

# Lista de todas las categorías esperadas
all_categoriesrtp = ['Acceptance Rate', 'Discipline Training', 'Distribution', 'Lack of Couriers']

# Encontrar categorías faltantes
missing_categoriesrtp = set(all_categoriesrtp) - set(dfrtp["SUB_CATEGORY"])

# Crear un DataFrame con las categorías faltantes
missing_datartp = pd.DataFrame({
    "SUB_CATEGORY": list(missing_categoriesrtp),
    "Cancel_rate": [0.0] * len(missing_categoriesrtp),  # Cancel_rate inicial como 0.0
    "Year": [2024] * len(missing_categoriesrtp),        # Año por defecto
    "Month": ["December"] * len(missing_categoriesrtp), # Mes por defecto
    "Day": [None] * len(missing_categoriesrtp)          # Día vacío
})

# Combinar con el DataFrame original
dfrtp = pd.concat([dfrtp, missing_datartp], ignore_index=True)

# Volver a formatear la columna Cancel_rate
dfrtp["Cancel_rate"] = dfrtp["Cancel_rate"].apply(lambda x: f"{x:,.2f}".replace(".", ",") + " %")

# Imprimir el DataFrame resultante
print(dfrtp)
```

```python
categories = dfrtp['SUB_CATEGORY'].unique()

# Luego, iteramos sobre cada día (clave del diccionario)
for day, day_df in dfs_by_day_rtp.items():
    # Obtenemos las categorías presentes para ese día
    day_categories = day_df['SUB_CATEGORY'].unique()
    
    # Creamos una lista de las categorías que faltan en ese día
    missing_categories = set(categories) - set(day_categories)
    
    # Si faltan categorías, las añadimos con Cancel_rate = 0
    for category in missing_categories:
        missing_row = pd.DataFrame({
            'SUB_CATEGORY': [category],
            'Cancel_rate': ['0,00 %'],
            'Year': [2024],        # Asegúrate de ajustarlo si es necesario
            'Month': ['December'], # Asegúrate de ajustarlo si es necesario
            'Day': [day]
        })
        # Añadimos las filas faltantes al DataFrame del día
        day_df = pd.concat([day_df, missing_row], ignore_index=True)
    
    # Ordenamos las filas por SUB_CATEGORY si deseas que esté en orden
    day_df = day_df.sort_values(by='SUB_CATEGORY').reset_index(drop=True)
    
    # Actualizamos el diccionario con el DataFrame modificado para ese día
    dfs_by_day_rtp[day] = day_df

# Imprimimos los DataFrames por día para verificar
print(dfs_by_day_rtp[16])
print(dfs_by_day_rtp[23])
```
